```python
def load_synonym_results(file_paths):
    results = {}
    for model_name, file_path in file_paths.items():
        df = pd.read_csv(file_path)
        synomym_count = df[df["prediction"]=="synonym"].shape[0]
        non_synonym_count = len(df) - synomym_count
        sim_scores = list(df["similarity"])

        row = {"number_of_synonyms": synomym_count, "number_of_non_synonyms": non_synonym_count, "similarity_scores": sim_scores}
        results[model_name] = row
    return results
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_similarity_scores(prefix, results, title_suffix="LEPIDOPTERA"):
    plt.figure(figsize=(12, 6))
    
    # 1. Use 'tab20' which contains 20 distinct colors (perfect for 12 models)
    palette = sns.color_palette("tab20", len(results))
    
    # 2. Sort results alphabetically so legend and color order match perfectly
    sorted_results = dict(sorted(results.items()))
    
    for i, (model_name, data) in enumerate(sorted_results.items()):
        sns.kdeplot(
            data["similarity_scores"], 
            label=model_name, 
            fill=True, 
            color=palette[i],  # Guaranteed unique color now
            alpha=0.35,        # Slightly lower alpha helps with heavy overlap
            linewidth=1.5
        )
        
    plt.title(f"Similarity Score Distributions for Synonyms ({title_suffix})", fontsize=14, fontweight='bold')
    plt.xlabel("Similarity Score", fontsize=12)
    plt.ylabel("Density", fontsize=12)
    
    # 3. Move legend outside or format it nicely so it doesn't block the data
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
    
    plt.tight_layout()
    plt.savefig(f"/Users/sefika/projects/lm_embeddings_4_species/figures/{prefix}_similarity_scores.pdf", dpi=300)
    plt.show()
```

```python
results = load_synonym_results(synonym_results_coleop)
for result in results.values():
    print(f"Number of synonyms: {result['number_of_synonyms']}, Number of non-synonyms: {result['number_of_non_synonyms']}")
plot_similarity_scores("coleop", results, title_suffix="COLEOPTERA")



```
